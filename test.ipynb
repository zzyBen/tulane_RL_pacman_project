```python
import json
import matplotlib.pyplot as plt
```

```python
def plot(version_1, version_2):
    with open(version_1) as json_file:
        data_1 = json.load(json_file)
    with open(version_2) as json_file:
        data_2 = json.load(json_file)

    fig, axs = plt.subplots(3)
    # Plot for version 1
    axs[0].plot(list(range(len(data_1['pacman_total_reward_list']))), data_1['pacman_total_reward_list'], label = "version 1", c='red')
    axs[1].plot(list(range(len(data_1['ghost_1_total_reward_list']))), data_1['ghost_1_total_reward_list'], label = "version 1", c='red')
    axs[2].plot(list(range(len(data_1['ghost_2_total_reward_list']))), data_1['ghost_2_total_reward_list'], label = "version 1", c='red')

    # Plot for version 2
    axs[0].plot(list(range(len(data_2['pacman_total_reward_list']))), data_2['pacman_total_reward_list'], label = "version 2", c='blue')
    axs[1].plot(list(range(len(data_2['ghost_1_total_reward_list']))), data_2['ghost_1_total_reward_list'], label = "version 2", c='blue')
    axs[2].plot(list(range(len(data_2['ghost_2_total_reward_list']))), data_2['ghost_2_total_reward_list'], label = "version 2", c='blue')

    plt.show()
```

```python

version_1 = 'output/2_cheating_pacman.json'
version_2 = 'output/3_cheating_pacman_no_ghost.json'
plot(version_1, version_2)
```
